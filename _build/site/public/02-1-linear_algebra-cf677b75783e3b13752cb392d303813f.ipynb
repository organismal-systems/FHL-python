## Creating matrices and higher-dimensional arrays

From a `numpy` perspective, a matrix is just a nested numpy array for which dimensions align. For example:

```python
A = np.array([[1, 4], [2, 3]])
print(A)
```

In principle there is no reason to stop at 2 indices, e.g.,

```python
M = np.array([[[1,2,3], [2,3,4]],[[-1, 0, 1], [-2, 0, 2]]])
print(M)
```

Two useful ways to build matrices are (1) start with 1D array and reshape it into 2 (or higher) dimensions; (2) use built-in matrix-creating functions

```python
B = np.arange(12).reshape(3, 4)
print(B)
```

```python
C = np.diag([1, 3, 5, 7])
print(C)
```

Part of the matrix can be modified via indexed assignment

```python
C[2][1] = -1
print(C)
```

## Eigenvalues and eigenvectors

There are four functions associated with matrix eigensystem:

+ `nla.eigvals()` solves the eigenvalues of a general matrix
+ `nla.eigvalsh()` solves the eigenvalues of a symmetric matrix
+ `nla.eig()` solves both eigenvalues and eigenvectors of a general matrix
+ `nla.eigh()` solves the eigenvalues and eigenvecto of a symmetric matrix

```python
# eigensystem of a matrix
A = np.array([[1, 4], [2, 3]])
D, V = nla.eig(A)
```

```python
print(D)
print(V)
```

```python
V @ np.diag(D) @ nla.inv(V)
```

```python
A @ V[:, 1]
```

```python
D[1] * V[:, 1]
```

## Cumulative exercise: Leslie matrix

The Leslie matrix describes how population structure changes with time. In one formulation, the matrix takes the general form:
\begin{bmatrix}
P_1 & F_1 & F_2 & \cdots & F_{N-1} \\
G_1 & P_2 & 0   & \cdots & 0 \\
0   & G_2 & P_3   & \cdots & 0 \\
\vdots & \vdots & \vdots & & \vdots \\
0 & \cdots & 0 & G_{N-1} & P_N
\end{bmatrix}

Given a population structure (counts of individual per age group) $\mathbf{n}_k = [n_{1,k}, n_{2,k}, \cdots, n_{N,k}]^T$ at time step $t(k)$, the population structure at time step $t(k+1)$ is given by $\mathbf{n}_{k+1} = L \, @ \, \mathbf{n}_k$

For concreteness, we will use the sea turtles data from [Crouse et.al., _Ecology_, 68(5), 1987](https://doi.org/10.2307/1939225), where, in the span of a year,

+ $\mathbf{P}$ = [0, 0.7370, 0.6610, 0.6907, 0, 0, 0.8089]
+ $\mathbf{G}$ = [0.6757, 0.0486, 0.0147, 0.0518, 0.8091, 0.8091]
+ $\mathbf{F}$ = [0, 0, 0, 0, 127, 4, 80] 

Where the stages represents:

```python
stages = ["eggs, hatchlings", "small juveniles", "large juveniles", "subadults", 
          "novice breeders", "1st-yr remigrants", "mature breeders"]
```

```python
# Function to build a Leslie matrix given per-age-class birthrate and per-age-class mortality

def make_Leslie_Matrix(F, P, G):
    M = np.diag(P) + np.diag(G, -1)
    M[0, 1:] = F
    return M
```

```python
F = [0, 0, 0, 127, 4, 80]
P = [0, 0.7370, 0.6610, 0.6907, 0, 0, 0.8089]
G = [0.6757, 0.0486, 0.0147, 0.0518, 0.8091, 0.8091]

L = make_Leslie_Matrix(F, P, G)
print(L)
```

# Visualizations in python

## The `matplotlib` package

To make plots in python we need a third-party package. The most common package for such purpose is `matplotlib`

```python
import matplotlib.pyplot as plt
```

In addition, we will use `numpy` to generate theoretical data and use `pandas` to load observed data:

```python
import numpy as np
import pandas as pd
```

## Example: a simple plot

Let's generate arrays that represent the $x$ and $y$ coordiantes of the function $y = \sqrt{x}$:

```python
x_array = np.linspace(0, 1, 100)
y_array = np.sqrt(x_array)
```

```python
x_array
```

```python
y_array
```

Let's make a line plot of the square root function.

`matplotlib` provides two plotting interfaces: one using the global `plt` object and the other using instances of `Figure` and `Axes` objects.

The `plt` interface is less verbose, but the local objects interface is more feature-complete.

```python
# using the global `plt` interface
plt.plot(x_array, y_array)
plt.show()
```

```python
# using the local object interface
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_array, y_array)
plt.show(fig)
```

## Adding plot elements

![An illustration showing how different parts of a figure corresponds to objects in the object-oriented plotting interface](img/sphx_glr_anatomy_001.webp)

**Figure**: The correpondence between plot elements and their object-oriented representations in `matplotlib`. Adopted from `matplotlib` [showcase](https://matplotlib.org/stable/gallery/showcase/anatomy.html).

Completing our previous plot by adding plot elements and specifying styles:

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x_array, y_array, lw=2, ls="--", c="tab:purple")

ax.set_xlim([0, 1])
ax.set_ylim([0, 1])

ax.grid(ls=":")

ax.set_xticks(np.arange(0, 1.05, 0.25))
ax.set_yticks(np.arange(0, 1.05, 0.25))

ax.set_xticks(np.arange(0, 1.01, 0.05), minor=True)
ax.set_yticks(np.arange(0, 1.01, 0.05), minor=True)
ax.tick_params("both", labelsize=11)

ax.set_title("The square root function", fontsize=14)
ax.set_xlabel("x-coordinate", fontsize=12)
ax.set_ylabel("y-coordinate", fontsize=12)

plt.show(fig)
```

## Plot types

```python
# scatter plot

x_array = np.random.normal(0, 1, 100)
y_array = 0.2 * x_array + 0.8 * np.random.normal(0, 1, 100)

fig = plt.figure()
ax = fig.add_subplot()

ax.scatter(x_array, y_array, marker=".")

plt.show(fig)
```

```python
# error bar plot

x_data = np.arange(0, 5.1, 0.5)
y_data = np.sqrt(x_data)
y_err = np.array([
    [0.4, 0.3, 0.3, 0.2, 0.4, 0.4, 0.3, 0.2, 0.2, 0.3, 0.3], # lower
    [0.1, 0.2, 0.2, 0.1, 0.2, 0.2, 0.1, 0.1, 0.2, 0.2, 0.2] # Upper
])
x_err = 0.1

fig = plt.figure()
ax = fig.add_subplot()

plt.errorbar(
    x_data, y_data, yerr = y_err, xerr = x_err, 
    color="blue", marker="o", lw=1, ls="--", 
    ecolor="black", capsize=3.0
)

plt.show(fig)

```

```python
# bar plot with error bars
# different color for each bar

x_data = np.arange(0, 5)
y_data = np.array([6, 5, 9, 4, 7])

y_err = np.array([0.7, 0.4, 0.3, 0.5, 0.2])

fig = plt.figure()
ax = fig.add_subplot()

ax.bar(
    x_data, y_data, align="center", width=0.8,
    tick_label = ["apple", "banana", "lemon", "grape", "orange"],
    color=["tab:blue", "tab:orange", "tab:green", "tab:red", "tab:purple"],
    yerr = y_err, ecolor="black", capsize=5
)

plt.show(fig)

```

```python
# 2D false color plot

X = np.linspace(-2, 2, 201)
Y = np.linspace(-1, 1, 101)

X_grid, Y_grid = np.meshgrid(X, Y)
Z = np.exp(-X_grid**2 - 0.5 * Y_grid**2)

fig = plt.figure()
ax = fig.add_subplot()

mesh = ax.pcolormesh(X, Y, Z, cmap="magma")
cb = fig.colorbar(mesh, aspect=15)

plt.show()
```

## Plotting from data file: body mass&ndash;metabolism allometry

**Data source**: White et. al., [_The scaling and temperature dependence of vertebrate metabolism_](https://doi.org/10.1098/rsbl.2005.0378)


```python
# loading dataset using pandas

df = pd.read_csv("data/allometry.csv", index_col=None)
df.info()
```

```python
# compute regression coefficients for the power law

from scipy.stats import linregress
fit = linregress(np.log(df["AvgMass"]), np.log(df["Q10SMR"]))
print(fit.slope, fit.intercept)
```

```python
# generate best fit curve

Mass_th = np.geomspace(np.min(df["AvgMass"]), np.max(df["AvgMass"]), 100)
Q10_th = np.exp(fit.intercept + fit.slope * np.log(Mass_th))
```

```python
# Plot the data and fit on log-log scale

fig = plt.figure()
ax = fig.add_subplot()

ax.scatter(df["AvgMass"], df["Q10SMR"], marker=".", c="tab:blue")
ax.plot(Mass_th, Q10_th, lw=2, ls="--", c="tab:orange")

ax.set_title("Body mass–metabolism allometry")
ax.set_xlabel("Body mass (g)")
ax.set_ylabel("SMR (mL O$_2$ h$^{-1}$)")

ax.set_xscale("log", base=10)
ax.set_yscale("log", base=10)

# save the figure to image file
#fig.savefig("allometry.png")

plt.show(fig)
```

# Basic jupyter notebook and python opeartions

## Jupyter notebook

+ A Jupyter notebook is comprised of **cells**
+ There are two main type of cells: markdown (for rich-text display) and code (for executing python code)
+ There are two modes of operation when working with a Jupyter notebook: the **command mode** and the **edit mode**. The former is for cell-level operations while the latter is for editing the content within a cell.
+ To run the codes in the current cell, press `Shift` + `Enter` in either edit mode or command mode. You can also use the play button (`⯈`) on the top of the main panel to execute the current cell.
  + `Shift` + `Enter` can also be used to render the current markdown cell, if it is not already rendered.
+ Some useful operations in command mode:
  + `m` converts the current cell into a markdown cell
  + `y` converts the current cell into a code cell
  + `a` inserts a cell above the current cell
  + `b` inserts a cell below the current cell
  + `c` copies the content of the current cell, and `v` pastes it
  + `d` + `d` deletes the current cell
  + `z` undoes the previous action
+ Some useful markdown formatting:
  + To create a list, start the line with a `+`, followed by a space
  + To set a block of texts in boldface, surround the block with a pair of `**`
  + To italicize a block of texts, surround the block with a pair of `_`
  + To set a block of text in monospace, surround the block with a pair of `` ` ``
  + To create headings, start the line with a sequence of `#`, followed by a space (the more `#` the lower the heading in the hierarchy)
+ **Important**: The python kernel remembers the order of code execution but _not_ the order of code presentation!

### Control flow

```python
# Branching

x = -20

if x > 0:
    abs_x = x
elif x == 0:
    abs_x = 0
else:
    abs_x = -x

abs_x
```

```python
# Looping

values = [1, 2, 3, 4]

total = 0
for x in values:
    total += x

total
```

```python
# Looping using the range object

total = 0
for x in range(1, 5):
    total += x

total
```

### Cumulative example: discrete logistic growth

Consider a population that follows the logistic growth model:

$$  p_{n+1} = p_{n} + r p_{n} \left( 1 - \frac{p_{n}}{K} \right)  $$

Write a function that takes $r$, $K$, $p_{1}$, and $N$ as input, and produce list of the size of the population from the first (given) to the $N$-th generation

```python
def logistic(r, K, p1, N):
    
    out = [ p1 ]
    p = p1
    
    for _i in range(N - 1):
        p = p + r * p * (1 - p/K)
        out.append(p)
        
    return out
```
